```python
url_gt = 'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet'
url_tz = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv'
```

```python
df_gt = pd.read_parquet(url_gt)
```

```python

dtype_tz = {
    "LocationID": "Int64",
    "Borough": "object",
    "Zone": "object",
    "service_zone": "object"
}
df_tz = pd.read_csv(url_tz, dtype = dtype_tz)
df_tz.shape
```

```python
from sqlalchemy import create_engine

engine = create_engine(f'postgresql://root:root@localhost:5432/ny_taxi')
df_gt.to_sql(
name='green_taxi_data',
con=engine,
if_exists='append'
)
```
